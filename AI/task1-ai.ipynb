```python
import re
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
```

```python
stemmer = PorterStemmer()
```

```python
def chatbot_response(user_input):
    # Preprocess the input
    user_input = user_input.lower()
    tokens = word_tokenize(user_input)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    # Define responses based on patterns
    if any(re.search(r'\bhi\b|\bhello\b', word) for word in stemmed_tokens):
        return "Hello! How can I assist you today?"
    elif any(re.search(r'\bhow\b.*\byou\b', word) for word in stemmed_tokens):
        return "I'm here to help you. How can I assist?"
    elif any(re.search(r'\bname\b', word) for word in stemmed_tokens):
        return "I'm a friendly chatbot created to assist you!"
    elif any(re.search(r'\bbye\b|\bexit\b', word) for word in stemmed_tokens):
        return "Goodbye! Feel free to return if you need more help."
    elif any(re.search(r'\bthank\b', word) for word in stemmed_tokens):
        return "You're very welcome!"
    elif any(re.search(r'\bhelp\b', word) for word in stemmed_tokens):
        return "I can answer questions or have a chat with you. What would you like to know?"
    else:
        return "I'm not sure I understand. Could you clarify?"
```

```python
print("Chatbot: Hello! Type 'exit' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Take care!")
        break
    print("Chatbot:", chatbot_response(user_input))
```
